```python
# Importing necessary libraries 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# Loading dataset
file_path = r'C:\Users\PRINCE\OneDrive\Documents\GitHub\Azubi_2\my2024-fuel-consumption-ratings.csv'
df = pd.read_csv(file_path, encoding='ISO-8859-1')

# Display the first few rows
df.head()
```

```python
# Display dataset information
df.info()
df.describe()
```

```python
# Renaming selected colums
df = df.rename(columns = {
    'Engine size (L)':'Engine_Size (L)',
    'Cylinders':'Cylinder',
    'Fuel consumption(L/100 km)':'Fuel_Consumption(L/100 km)',
    'CO2 emissions (g/km)': 'CO2_Emissions (g/km)'
    })
df.head()
```

```python
# Displaying selected columns
selected_columns = df[['Engine_Size (L)', 'Cylinder', 'Fuel_Consumption(L/100 km)', 'CO2_Emissions (g/km)']]
selected_columns
```

```python
# Target
y = selected_columns['CO2_Emissions (g/km)']

# Feature
x = selected_columns['Engine_Size (L)']
```

```python
# Import necessary libraries
from sklearn.model_selection import train_test_split

# Splitting dataset
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
# Importing model
from sklearn.linear_model import LinearRegression

# Initializing linear regression model
model = LinearRegression()

# Ensure x (features) is in 2D format
x_train = x_train.values.reshape(-1, 1)
x_test = x_test.values.reshape(-1, 1)

# Training the model on the training data
model.fit(x_train, y_train)
```

```python
# Make predictions on the test data
y_predicted = model.predict(x_test)
y_predicted
```

```python
# Evaluate the model
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_predicted)
mse
```

```python
rsquare = r2_score(y_test, y_predicted)
rsquare
```

```python
rmse = np.sqrt(mse)
rmse
```
