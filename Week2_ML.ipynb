### Skewness is a measure of the asymmetry of the data distribution

* Skewness = 0: Perfectly symmetrical distribution(like a normal distribution)
* Skewness > 0: Positive skew(right-skew) - the right tail is long or flatter
* Skewness < 0: Left skew(left-skew) - the left tail is long or flatter


```python
# Checking  dataset for skewness
selected_columns.skew()
```

##### Featuring Engine Size(L) as an Independent Variable

```python
# Target
y = selected_columns['CO2_Emissions (g/km)']

# Feature
x = selected_columns['Engine_Size (L)']
```

```python
# Importing necessary libraries
from sklearn.model_selection import train_test_split

# Splitting dataset
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
# Importing model
from sklearn.linear_model import LinearRegression

# Initializing linear regression model
model = LinearRegression()

# Ensure x (features) is in 2D format
x_train = x_train.values.reshape(-1, 1)
x_test = x_test.values.reshape(-1, 1)

# Training the model on the training data
model.fit(x_train, y_train)
```

```python
# Make predictions on the test data
y_predicted = model.predict(x_test)
y_predicted
```

```python
# Evaluating the model
from sklearn.metrics import mean_squared_error, r2_score

# Calculating Mean Squared Error (MSE) between actual and predicted values
mse = mean_squared_error(y_test, y_predicted)
mse
```

```python
# Calculating R-squared
rsquare = r2_score(y_test, y_predicted)
rsquare
```

```python
# Calculating Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)
rmse
```
