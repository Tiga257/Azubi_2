#### Regrouping Stages Column

```python
# Import the regular expression module
import re

# Defining a function to categorize funding stages
def funding_stage(full_data, column_name='Stage'):
    # Definining a dictionary to map patterns to categories
    stages = {
        # Preseed stage patterns
        'Preseed': [r'\bpre[-]?seed\b'],
        # Seed stage patterns
        'Seed': [r'\bseed|\bangel\b', r'\bseed funding\b', r'\bseed round\b', r'\bseed\+|\bseed fund'],    
        # Pre Series stage patterns
        'Pre Series': [r'\bpre[-]?series\b'],
        # Series A stage patterns
        'Series A': [r'\bseries[-]?a\b', r'\bseries a\b'],
        # Series B stage patterns
        'Series B': [r'\bseries[-]?b\b'],  
        # Series C stage patterns
        'Series C': [r'\bseries[-]?c\b'],
        # Later Stage patterns
        'Later Stage': [r'\bseries[-]?[defghij]\b', r'blater stage\b', r'\bseries\b'], 
        # Equity patterns
        'Equity': [r'\bpe\b', r'\bequity\b', r'\bprivate equity\b', r'\bgrant\b', r'\bdebt\b', r'bpost-ipo\b', r'\bcorporate round\non-equity\b'], 
        # Undisclosed patterns
        'Undisclosed': [r'\bundisclosed\b', r'\bunknown\b', r'\bventure - series unknown\b']
    }

    # Initialize an empty list to store categorized stages
    categorized_stages = []

    # Iterate over each stage in the input data
    for stage in full_data[column_name]:
        # Convert the stage to lowercase for case-insensitive matching
        stage = stage.lower()
        
        # Find the category that matches the stage
        category = next((cat for cat, patterns in stages.items() if any(re.search(pattern, stage) for pattern in patterns)), 'Others')
        
        # Append the category to the list
        categorized_stages.append(category)

    # Return the list of categorized stages
    return categorized_stages

# Apply the function to the input data
full_data['Stage'] = funding_stage(full_data, column_name='Stage')
```

## Hypothesis Testing

Null: Funding amount received by tech startups does not significantly differ from non-tech startups funding received.

Alternative: Funding amount received by tech startups does differ from non-tech startups funding received.

```python
# Filtering the full_data dataframe to include only rows where 'Tech_Status' is 'Tech'
tech_data = full_data.loc[full_data['Tech_Status'] == 'Tech']

# Displaying the filtered dataframe
tech_data
```

```python
# Filtering the full_data dataframe to include only rows where 'Tech_Status' is not 'Tech'
non_tech_data = full_data.loc[full_data['Tech_Status'] != 'Tech']

# Display ingthe filtered dataframe
non_tech_data
```

#### Mann-Whitney U Test for Difference in Funding Amount between Tech and Non-Tech Sector

```python
from scipy.stats import mannwhitneyu

# Performing the Mann-Whitney U test
statistic, p_value = mannwhitneyu(tech_data['Amount($)'], non_tech_data['Amount($)'])

# Output results
print('Mann-Whitney U statistic', statistic)
print('p_value', p_value)

# At an alpha level of 0.05
alpha=0.05

if p_value<alpha:
    print('We do not accept the null hypothesis')
else:
    print('We do accept the null hypothesis')
```

#### Detailed analysis

```python
# Importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Perform Mann-Whitney U test
u_stat, p_value = mannwhitneyu(non_tech_data['Amount($)'], tech_data['Amount($)'])

# Print results
print(f"Mean Amount($) non-tech: {non_tech_data['Amount($)'].mean():.2f}")
print(f"Mean Amount($) tech: {tech_data['Amount($)'].mean():.2f}")
print(f"p-value: {p_value:.3f}")

# Create bar chart
plt.bar(['Non-Tech', 'Tech'], [non_tech_data['Amount($)'].mean(), tech_data['Amount($)'].mean()], color='black', edgecolor='red')
plt.ylabel('Amount($)')
plt.title('Funding Comparison')
plt.show()

# Create box plot
combined_data = pd.concat([non_tech_data.assign(Group='Non-Tech'), tech_data.assign(Group='Tech')])
sns.boxplot(x='Group', y='Amount($)', data=combined_data)
plt.title('Funding Amounts Box Plot')
plt.show()

# Create violin plot
sns.violinplot(x='Group', y='Amount($)', data=combined_data)
plt.title('Funding Amounts Violin Plot')

# Show the plot
plt.show()
print()
```

##### Based on the analysis, we can conclude the following:

#### Reject the Null Hypothesis

##### The p-value (0.001) is less than the significance level (0.05), indicating that there is a statistically significant difference in the funding amounts between tech and non-tech startups.

#### Alternative Hypothesis is Supported

##### The mean funding amount for tech startups is significantly higher than the mean funding amount for non-tech startups.

#### Conclusion

##### The analysis suggests that tech startups receive significantly more funding than non-tech startups. This supports the alternative hypothesis and indicates a difference in funding patterns between the two groups.

### Q4. How many sectors have received funding from multiple investors 

```python
# Import the necessary libraries
import pandas as pd

# Filtering the data to include only startups with multiple investors (more than 1 comma)
multiple_investors_data = full_data[full_data['Investor'].str.count(',') > 0]

# Filtering the data to include only startups with single investors (0 or 1 comma)
single_investors_data = full_data[full_data['Investor'].str.count(',') == 0]

# Counting the number of sectors with multiple investors
sectors_with_multiple_investors = len(multiple_investors_data['Sector'].unique())

# Counting the number of sectors with single investors
sectors_with_single_investors = len(single_investors_data['Sector'].unique())

# Printing the result
print()
print("1. The number of sectors with multiple investors is:", sectors_with_multiple_investors)
print()
print("2. The number of sectors with single investors is:", sectors_with_single_investors)
print()

# Create a pie chart to visualize the proportion of sectors with multiple investors
plt.figure(figsize=(8,8))
plt.pie([sectors_with_multiple_investors, sectors_with_single_investors], labels=['Multiple Investors', 'Single Investor'], autopct='%1.1f%%')
plt.title('Proportion of Sectors with Multiple Investors')

# Show the plot
plt.show()
```

### Q6. What are the top 7 Company Brands with the highest growth rate in funding.

```python
# Calculating the growth rate for each company
growth_rates = full_data.groupby('Company_Brand')['Amount($)'].apply(lambda x: (x.max() - x.min()) / x.min())/100

# Sortting the growth rates in descending order and select the top 10 companies
top_7_companies = growth_rates.sort_values(ascending=False).head(7)

# Printing the result
print("\nThe top 7 companies with the highest growth rate are:")
print(top_7_companies)  
print()

# Creating a pie chart
plt.figure(figsize=(10,8))
plt.pie(top_7_companies, labels = top_7_companies.index, autopct='%1.1f%%')
plt.title('Growth Rate by Company (Top 7)')

# Show the plot
plt.show()
```
